```python
response = requests.get('https://www.instagram.com/explore/locations/212988663/new-york-new-york/?__a=1')
new_york = response.json()
new_york = pd.read_json(json.dumps(new_york['location']['top_posts']['nodes']))

response = requests.get('https://www.instagram.com/explore/locations/6889842/paris-france/?__a=1')
paris = response.json()
paris = pd.read_json(json.dumps(paris['location']['top_posts']['nodes']))

response = requests.get('https://www.instagram.com/explore/locations/31499759/rome-italy/?__a=1')
rome = response.json()
rome = pd.read_json(json.dumps(rome['location']['top_posts']['nodes']))

response = requests.get('https://www.instagram.com/explore/locations/213131048/berlin-germany/?__a=1')
berlin = response.json()
berlin = pd.read_json(json.dumps(berlin['location']['top_posts']['nodes']))

response = requests.get('https://www.instagram.com/explore/locations/213262633/delhi/?__a=1')
delhi = response.json()
delhi = pd.read_json(json.dumps(delhi['location']['top_posts']['nodes']))

response = requests.get('https://www.instagram.com/explore/locations/214179276/minsk-belarus/?__a=1')
minsk = response.json()
minsk = pd.read_json(json.dumps(minsk['location']['top_posts']['nodes']))
```
